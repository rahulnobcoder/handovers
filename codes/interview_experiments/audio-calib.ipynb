```python
from gtts import gTTS
import zipfile
import os

sentences = [
    "For some people, this room might be",
    "the scariest place on Earth.",
    "Behind these black curtains are deadly spiders.",
    "Hundreds of them.",
    "And what we're gonna do is poke them, make them angry,",
    "and then suction the venom that appears",
    "at the end of their really long fangs",
    "This is about as close as I ever want to get to a funnel.",
    "And we're doing it for a very good reason.",
    "This is a funnel web spider.",

]

output_dir = "audio_files"
os.makedirs(output_dir, exist_ok=True)

audio_files = []
for i, sentence in enumerate(sentences):
    audio_path = os.path.join(output_dir, f"sentence_{i+1}.mp3")
    tts = gTTS(sentence, lang='en', tld='co.in')  # Indian accent
    tts.save(audio_path)
    audio_files.append(audio_path)
print(len(audio_files))

zip_path = "indian_accent_audio.zip"
with zipfile.ZipFile(zip_path, 'w') as zipf:
    for file in audio_files:
        zipf.write(file, os.path.basename(file))

print(f"Zip file created: {zip_path}")
```

```python
import boto3
from zipfile import ZipFile
import os

# Define the text for each line as per the image
lines= [
    "For some people, this room might be",
    "the scariest place on Earth."
    "Behind these black curtains are deadly spiders.",
    "Hundreds of them.",
    "And what we're gonna do is poke them, make them angry,",
    "and then suction the venom that appears",
    "at the end of their really long fangs"
    "This is about as close as I ever want to get to a funnel.",
    "And we're doing it for a very good reason.",
    "This is a funnel web spider.",

]

# List of voices to cycle through
voice_names = ["Aditi", "Raveena"]  # Amazon Polly Indian English voices

# Initialize the Polly client
polly = boto3.client("polly", region_name="us-west-2")  # Specify your desired region

# Folder to store audio files
audio_folder = "audio_files"
os.makedirs(audio_folder, exist_ok=True)

# Generate audio files with different voices
for i, line in enumerate(lines):
    voice_name = voice_names[i % len(voice_names)]  # Cycle through voices

    response = polly.synthesize_speech(
        Text=line,
        OutputFormat="mp3",
        VoiceId=voice_name,
        LanguageCode="en-IN"
    )

    # Save the output
    audio_path = os.path.join(audio_folder, f"line_{i+1}.mp3")
    with open(audio_path, "wb") as file:
        file.write(response["AudioStream"].read())

# Create a zip file containing all audio files
zip_path = "audio_files.zip"
with ZipFile(zip_path, 'w') as zipf:
    for root, _, files in os.walk(audio_folder):
        for file in files:
            zipf.write(os.path.join(root, file), os.path.relpath(os.path.join(root, file), audio_folder))

print(f"Audio files with different voices are saved and zipped as {zip_path}")
```
